```python
import numpy as np
```

```python
vector_1 = np.array([1,2,3])
vector_2 = np.array([4,5,6])
vector_1
```

Dot

```python
dot_product = np.dot(vector_1, vector_2)
dot_product
```

```python
vector_1_norm = np.linalg.norm(vector_1)
vector_2_norm = np.linalg.norm(vector_2)
vector_1_norm
```

```python
cosine_similaraty = dot_product / (vector_1_norm * vector_2_norm)
cosine_similaraty
```

```python
matris1 = [1 , 5, 8], [3, 7, 11]
matrix_multiplication = matris1 @ vector_1
matrix_multiplication
```

```python
def f(x):
    return x**3 + 2*x
f(13)
```

```python
import torch
x = torch.tensor(13.0)
y = f(x)
print(y)
```

```python
if torch.cuda.is_available():
    # This handles both NVIDIA (CUDA) and AMD (ROCm) on Linux/Windows
    device = "cuda"
    backend = "ROCm" if torch.version.hip else "CUDA"
elif torch.backends.mps.is_available():
    # This handles Apple Silicon and supported AMD GPUs on macOS
    device = "mps"
    backend = "MPS"
else:
    device = "cpu"
    backend = "CPU"

print(f"Using device: {device} ({backend} backend)")
```

```python
from sklearn.datasets import load_iris
data = load_iris()

X = data["data"]
y = data["target"]
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=3000)

model.fit(X,y)

print(model.score(X,y))
```
